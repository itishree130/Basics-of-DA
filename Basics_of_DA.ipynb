 4. What is Prompt Engineering, and explain any two types of prompts used in Generative AI?

 ->  Prompt Engineering is the process of designing and refining the instructions (prompts) given to an AI model to get accurate, relevant, and useful responses. It helps users communicate clearly with AI tools like ChatGPT or image generators to achieve better results.

 Two common types of prompts used in Generative AI are:

Instruction Prompt:

Tells the AI exactly what to do.

Example: “Summarize this paragraph in three sentences.”

Use: Helps get focused and task-specific responses.

Few-Shot Prompt:

Provides examples along with the task to guide the AI’s response.

Example:

“Translate the following words to French:

Dog → Chien

Cat → Chat

Bird → ?”

Use: Improves accuracy when AI sees examples before performing a similar task.

8. Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.

-> Here’s a clear explanation of the evolution of analytics, showing how it has advanced over time:

1> Descriptive Analytics – “What happened?”

Focuses on summarizing past data using reports, dashboards, and charts.

Example: Monthly sales reports or website traffic summaries.

2> Diagnostic Analytics – “Why did it happen?”

Examines data to find causes behind trends or problems.

Example: Identifying why sales dropped in a specific region.

3> Predictive Analytics – “What is likely to happen?”

Uses statistical models and machine learning to forecast future outcomes.

Example: Predicting customer churn or product demand.

4> Prescriptive Analytics – “What should we do about it?”

Suggests the best actions or decisions using optimization and AI models.

Example: Recommending pricing strategies or marketing actions.

5> Generative AI Analytics – “What can AI create or automate?”

Uses AI models (like ChatGPT) to generate insights, reports, visualizations, and even predictive models automatically.

Example: AI tools creating instant business reports or suggesting new business ideas from data.
